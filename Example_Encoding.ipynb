### Define Model

```python
x_input = Input(shape=(len(x_data.columns),), name='input')
x = Dense(516, activation='relu')(x_input)
x = Dense(516, activation='relu')(x)
x_output = Dense(len(y_data.columns), activation='sigmoid')(x)
```

```python
model = Model(inputs=[x_input], outputs=[x_output])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
hist = model.fit(x_data, y_data, epochs=50, batch_size=64, validation_split=0.2)
```
